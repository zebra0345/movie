```python
# 영화 1 : 영화의 평점이 10/10 -> 5명 평가
# 영화 2 : 영화 평점 8/10 -> 100명 평가
```

```python
# 평점 가중치부여단계

# 평점 평균
C = df2['vote_average'].mean()

# 상위 10프로에 대한 데이터를 추출(하위 0.9프로)
m = df2['vote_count'].quantile(0.9)
# m 보다 많은 데이터를 필터링
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
# 최소 1840개 이상 데이터
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m)*R) + (m/(m+v)*C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head(3)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score']].head()
```

```python
pop= df2.sort_values('popularity', ascending=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

plt.barh(pop['title'].head(6),pop['popularity'].head(6), align='center',
        color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.title("Popular Movies")
```

### 2. Content based 필터링

```python
# 원본데이터
df2['overview'].head(5)
```

```python
# 텍스트 == 백터화
# Bag Of Wards - BOW : 문서 텍스트에서 단어가 몇번이나 나왔나를 분석
# 피쳐 백터화 : 각 피쳐에(단어에) 카운트 할당

# -- 문서 100개 모든 문서에서 나온 단어 10000개
# 피쳐 백터화를 하면 => 100 * 10000 = 100만개 요소가 나온다.

# 1. TfidVectorizer (overview : 관사같은거 제외 - the, a 등)
# 2. CountVectorizer (있는 그대로 단어들을 센다.)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
ENGLISH_STOP_WORDS
```

```python
df2['overview'].isnull().values.any()
```

```python
# 공백 제거
df2['overview'] = df2['overview'].fillna("")
```

```python
# 백터화
tfidf_matrix = tfidf.fit_transform(df2['overview'])
tfidf_matrix.shape
```

```python
# 코사인 유사도
from sklearn.metrics.pairwise import linear_kernel

cosine_sin = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sin

# 같은 문장은 다 1, 나머지는 1을 기준으로 대칭되는 구조
```

```python
cosine_sin.shape
```

```python
# 영화제목을 입력하면 인덱스를 알려주는 함수
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
```

```python
indices['The Dark Knight Rises']
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해 상위 10개 영화목록반환

def get_recommand(title, cosin_sin = cosine_sin):
    idx = indices[title]
    
    # 코사인 유사도를 인덱스와 쌍으로 가져오기
    sin_score = list(enumerate(cosine_sin[idx]))
    
    # 코사인유사도 기준 내림차순정렬
    sim_scores = sorted(sin_score, key=lambda x:x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    
    # 추천목록의 10개 인덱스 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보로 영화제목
    return df2['title'].iloc[movie_indices]
```

```python
cosine_sin[3]
```

```python
# 코사인 유사도를 인덱스와 쌍으로 가져오기
test_sin_score = list(enumerate(cosine_sin[3]))
```

```python
sorted(test_sin_score, key=lambda x:x[1], reverse=True) # 코사인유사도 기준 내림차순정렬
test_sin_score[1:11]
```

### 다양한 요소 기반 추천 (장르, 감독, 키워드)

```python
df2.loc[0, 'genres']
```

```python
# str => list 전환
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']

for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
df2.loc[0, 'crew']
```

```python
# 감독정보만 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df2['director'] = df2['crew'].apply(get_director)
```

```python
df2[df2['director'].isnull()]
```

```python
# value 추출

def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']]
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])

df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
df2['overview']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
# 코사인유사도

from sklearn.metrics.pairwise import cosine_similarity

cosine_sin2 = cosine_similarity(count_matrix, count_matrix)
cosine_sin2
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
indices
```
